```python
import numpy as np
import rasterio as rio
import glob
```

```python
#Follwing shows the normalization values that has previousely been applied to the patches
# To covert tha pixel values to their original values, these coeficients are used
Norm1 = np.empty((10,224,224,8))
Norm2 = np.empty((10,224,224,8))

DMSP_Norm1 = (np.ones(((224,224,8)))*np.array([[[0.03512808462999218, 0.050729950158437294, 0.07239252056994558, 0.12444550248291769,
                                                 0.10850729467914068, 71.28094889057179, 0.09077465132496389,  18.566259343422725]]])).astype(np.float16)
DMSP_Norm2 = (np.ones(((224,224,8)))*np.array([[[0.06614169065743208, 0.09635988259340068, 0.11381173286380034, 0.2403052018705142,
                                                 0.169467635861788, 281.2976763110107, 0.24095368620456603,  13.069690484896771]]])).astype(np.float16)
VIIRS_Norm1 = (np.ones(((224,224,8)))*np.array([[[0.03512808462999218, 0.050729950158437294, 0.07239252056994558, 0.12444550248291769,
                                                  0.10850729467914068, 71.28094889057179, 0.09077465132496389, 15.861609457767129]]])).astype(np.float16)
VIIRS_Norm2 = (np.ones(((224,224,8)))*np.array([[[0.06614169065743208, 0.09635988259340068, 0.11381173286380034, 0.2403052018705142,
                                                  0.169467635861788, 281.2976763110107, 0.24095368620456603,  5.075756473790338]]])).astype(np.float16)
for i in range(8):
    Norm1[i,:,:,:] = DMSP_Norm1
    Norm2[i,:,:,:] = DMSP_Norm2
for i in range(2):
    Norm1[i+8,:,:,:] = VIIRS_Norm1
    Norm2[i+8,:,:,:] = VIIRS_Norm2    
```

```python
# load the patches from their directory
list_blobs = glob.glob('/mimer/NOBACKUP/groups/globalpoverty1/Kakooei/DatasetAfricaPoverty/Markus/TenYears/Geo41/*.tif') 
print(len(list_blobs))
```

```python
# funcOrg returns the original pixel value of each patch, as the stored patches are normalized by fixed values
def funcOrg(blob):
    img_multi = rio.open(blob)
    img_array = img_multi.read()
    imtransform = img_multi.transform
    nRows =  img_array.shape[1]
    nCols =  img_array.shape[2]
    Bands = img_array.shape[0]
    img_array_reshape = img_array.reshape(Bands, nRows*nCols).T
    img_array_reshape = img_array_reshape.reshape(nRows,nCols,Bands)
    cols, rows = np.meshgrid(np.arange(nCols), np.arange(nRows))
    xs, ys = rio.transform.xy(imtransform, rows, cols)
    lons = np.array(xs)
    lats = np.array(ys)
    
    img_array_reshape2 = img_array_reshape.reshape(10,nRows,nCols,8)
    for i in range(10):
        img_array_reshape2[i,:,:,:] = (np.add(np.multiply(img_array_reshape[:,:,i*8:(i+1)*8],
                                                            Norm1[i,:,:,:]),Norm2[i,:,:,:])).reshape(1,nRows,nCols,8)
    
    return img_array_reshape2, lons, lats
```

```python
ImgTemp, lons, lats = funcOrg(list_blobs[1])
```
